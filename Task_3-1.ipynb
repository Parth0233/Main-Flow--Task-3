```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('householdtask3.csv')
```

```python
data.head(21)
```

```python
#Scatter Plot with age against income

plt.scatter(data['age'], data['income'])

#Adding title to the plot
plt.title("SCATTER PLOT")

#Setting the x and y Labels
plt.xlabel('age')
plt.ylabel('income')

#Showing the chart
plt.show()
```

```python
#Line chart with year against own

plt.plot(data['year'])
plt.plot(data['own'])

#Adding title to the plot
plt.title("LINE CHART")

#Setting the x and y Labels
plt.xlabel('year')
plt.ylabel('own')

#Showing the chart
plt.show()
```

```python
#Bar chart on the bar plot

plt.bar(data['year'], data['expenditure'])

#Adding title to the plot
plt.title("BAR CHART")

#Setting the x and y Labels
plt.xlabel('year')
plt.ylabel('expenditure')

#Showing the chart
plt.show()
```

```python
#Histogram

plt.hist(data['own_prop'])
plt.title("Histogram of Own_prop")
plt.xlabel("own_prop")
plt.show()
```
